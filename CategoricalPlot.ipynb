# Categorical data plots

Now let's discuss with seaborn to plot categorical data! There are few main plot types for this:

countplot() 

barplot() 

boxplot() 

violinplot() 

stripplot()

swarmplot()

```python
import matplotlib.pyplot as pyplot
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

```python
df["sex"].value_counts()
```

#### countplot() 
sns.countplot() counts the number of occurrences of each category in a column and plots them as bars.

It is basically a bar chart for categorical data, but you don’t need to manually aggregate counts (Seaborn does it for you).

```python
sns.countplot(x=df['sex'])
```

```python
# if you want to check no of men and women who smoke and dont smoke 
df.groupby("smoker")["sex"].value_counts()
```

```python
sns.countplot(x=df['sex'],hue=df['smoker'])
```

### 🔹 What is barplot?

sns.barplot() shows the mean (average) value of a numerical variable for each category.

```python
sns.barplot(x=df['sex'],y=df["total_bill"])
```

```python
import numpy as np 

sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)
```

### A boxplot shows the distribution of numerical data across categories.

It summarizes data using 5 key statistics:

Minimum (excluding outliers)

Q1 (25th percentile)

Median (50th percentile)

Q3 (75th percentile)

Maximum (excluding outliers)

It also highlights outliers (points outside the whiskers).

👉 It’s sometimes called a "whisker plot".

```python
sns.boxenplot( x="tip",y="day",data=df, palette="rainbow")
```

### 🔹 What is a Violin Plot?

A violinplot combines a boxplot + KDE (Kernel Density Estimation).

The "violin" shape shows the distribution density of the data.

Inside, it also shows the median and quartiles (like boxplot).

👉 So it’s like:
Boxplot = summary (quartiles + outliers)
Violinplot = summary + shape of distribution

```python
sns.violinplot(x="day",y="total_bill",data=df, palette="rainbow")
```

## stripplot()
stripplot will create a scatterplot.

Plots individual data points along a categorical axis.

Points may overlap, making it hard to see density in large datasets.

```python
sns.stripplot(x="day",y="total_bill",data=df)
```

## swarmplot()

It is similar to stripplot but point are adjust so they dont overlap. An improved version of stripplot.

Uses a smart algorithm to avoid overlap of points.

Points are adjusted (like a swarm of bees 🐝), so every point is visible.

```python
sns.swarmplot(x="day",y="total_bill",data=df)
```

```python
# Createing combination of violinplot and swarmplot

sns.violinplot(x="tip",y="day",data=df)
sns.swarmplot(x="tip",y="day",data=df)
```
