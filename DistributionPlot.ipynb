### We use distribution plots to visualize the distribution of quantitative data.
### Lets discuss some plots that allow us to visualize the distribution of a data set. These plots are:
1) distplot/histplot
2) jointplot
3) pairplot
4) rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
df["size"].unique()
```

#### Histogram is best if you want to check distribution of your data


```python
sns.histplot(df["total_bill"],bins=20,kde=True)  #kde --> Kernel Density Estimation
```

```python
sns.kdeplot(df['total_bill'])    #KDE plot to show only Kernel Density Estimation
```

```python
# Now with the matplotlib we can make bot the histogram of "total_bill" & "tip" is single canvas
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"], bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins=20,kde=True)
```

#### Use jointplot if you want to chen distribution of two quantitative data

```python
sns.jointplot(x="total_bill",y="tip", data=df,kind="scatter")  #it will use scatter plot
```

```python
sns.jointplot(x='total_bill',y="tip",data=df,kind="kde", fill=True)
```

```python
sns.jointplot(x="total_bill",y='tip',data=df,kind="hist")
```

```python
sns.jointplot(x="total_bill",y='tip',data=df,kind="hex")
```

```python
sns.jointplot(x="total_bill",y='tip',data=df,kind="reg")
```

```python
sns.jointplot(x="total_bill",y='tip',data=df,kind="resid")
```

### sns.pairplot()

📌 What it does:

Creates a grid of plots for every numerical column vs every other numerical column.

Diagonal = histogram / KDE of each variable.

Off-diagonal = scatterplots (relationship between variables).

👉 It’s like a quick data exploration tool to check correlations.

```python
sns.pairplot(df,hue='sex')
```

```python
df
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

### What is rugplot?

A rug plot is the simplest plot:

It draws small vertical (or horizontal) lines for each data point along an axis.

It’s like dropping “ticks” on the axis wherever data exists.

```python
sns.rugplot(df)
```
