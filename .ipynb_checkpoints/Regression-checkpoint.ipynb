# lmplot
regression plot is used in machine learning, **lmplot()** we use to create it. Actually It’s mainly used for showing regression lines + scatter plots.
y-axis is where we put what we want to predict and x-axis is where we put on what data we want to predict.


```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
tps=sns.load_dataset("tips")
tps
```

```python
sns.lmplot(x="total_bill",y="tip",data=tps)

# in this we are predicting what will be tip based on total bills
```

```python
sns.lmplot(x="total_bill",y="tip",data=tps, hue="sex")
```

```python
sns.lmplot(x="total_bill",y="tip",data=tps, hue="sex", palette='coolwarm', markers=["v","o"])
```
